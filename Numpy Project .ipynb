```python
import numpy as np
print(np.__version__)
```

```python
print("Scores of the students")
marks = [45,67,89,76,55,90]
array_marks = np.array(marks)
print(array_marks)
highest_marks = np.max(array_marks)
lowest_marks = np.min(array_marks)
Avg_marks = np.mean(array_marks)

# Basic analysis
print("               BASIC ANALYSIS             ")
print(f"The highest marks in the class is {highest_marks}")
print(f"The lowest marks in the class is {lowest_marks}")
print(f"The Average marks of the class is {Avg_marks:.2f}")
```

```python
# Performance categories
print('             PERFORMANCE CATEGORIES                   ')
topper_student = array_marks[array_marks>80]
formatted_topperStudent = " , ".join ([f"{var}" for var in topper_student ])
Average_student = array_marks[(array_marks>=50)&(array_marks<80)]
formatted_avgStudent = " , ".join ([f"{var1}" for var1 in Average_student ])
Need_improvment = array_marks[array_marks<50]
formatted_weakStudent = " , ".join ([f"{var2}" for var2 in Need_improvment])

print(f"Congratulations! You really nailed it {formatted_topperStudent}")
print(f'Nice! Keep it up {formatted_avgStudent}')
print(f'You need improvement {formatted_weakStudent}')
```

```python
# Count of students in each category 
print("             COUNT OF EACH CATEGORY              ")
topper = np.sum(array_marks>=80)
Avg = np.sum((array_marks>=50)&(array_marks<80))
weak = np.sum(array_marks<=50)
total = array_marks.size

print(f"Count of toppers in the class is {topper}")
print(f"Count of Average students in the class is {Avg}")
print(f"Count of weak students in the class is {weak}")
```

```python
# Percentage of topper and weak student
print("                 PERCENTAGE OF EACH CATEGORY        ")
topper_percentage = (topper/total)*100
Avg_percentage = (Avg/total)*100
Weak_percentage = (weak/total)*100

print(f"The percentage of total toppers in the class is {topper_percentage:.2f} %")
print(f"The percentage of total Average students in the class is {Avg_percentage:.2f} %")
print(f"The percentage of total weak students in the class is {Weak_percentage:.2f} %")
```

```python

# Adding grace 
print("                    UPDATED SCORES               ")
updated_score = array_marks + 5
print(f"The updated marks of the students are {updated_score}")
```
