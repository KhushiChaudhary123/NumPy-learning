```python
import numpy as np
print(np.__version__)
```

```python
shop_sales = [120,150,180,200,170]
shop_array= np.array(shop_sales)
print(shop_array)
total = np.sum(shop_array)
print(f"Total sales:{total}")
Average= np.mean(shop_array)
print(f"Average sales:{Average}")
```

```python
student_marks = [45,67,89,76,55,90]
marks = np.array(student_marks)
print(marks)
highest = np.max(marks)
lowest = np.min(marks)
Avg = np.mean(marks)

print(f'The highest marks is : {highest}')
print(f'The lowest marks is : {lowest}')
print(f'The Average marks is : {Avg:.2f}')
```

```python
distance = [2,3,4,5,3,4,6]
distance2 = np.array(distance)
print(distance2)
Total_Distance= np.sum(distance2)
Greater_than_4 = distance2[distance2>4]
formatted_distances = " and ".join([f"{val} km" for val in Greater_than_4])
print(f'The total distance covered in 7 days is {Total_Distance} km')
print(f'The distance covered greater than 4 km in 7 days is {formatted_distances}')
```

```python
prices_of_items = [[10,20,30],
                   [15,25,35]]
prices_of_shops = np.array(prices_of_items)
print(prices_of_shops)
# by using slicing 
print(f'The 3rd item of the 2nd shop is {prices_of_shops[1,2]}')

Total_price = np.sum(prices_of_shops)
print(f'The total price of both shop item is {Total_price}')
```
